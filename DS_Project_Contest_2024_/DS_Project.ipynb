## พิจารณาจังหวัด แพร่

```python
phrae = phrae_maeh.loc[phrae_maeh.re_PROVINCE.str.contains("แพร่", na=False)]
phrae
```

เนื่องจากทุก re_AMPHUR ที่เป็น Missing Value re_TAMBON and re_ZIPCODE จะเป็น Missing Value ด้วย และไม่ได้นำมาประกอบการพิจารณาจึงตัว Column ทิ้งได้เลย ใน จังหวัดแพร่

```python
phrae = phrae.drop(["re_TAMBON", "re_ZIPCODE"],axis=1)
phrae
```

```python
phrae["re_AMPHUR"].isnull().sum()
```

```python
# for predictoin
amphur_miss_ph = phrae.loc[phrae['re_AMPHUR'].isnull()]
amphur_miss_ph
```

```python
one_hot_d = pd.get_dummies(amphur_miss_ph['DEST'], prefix='', prefix_sep='',dtype=int)

amphur_miss_ph = amphur_miss_ph.drop(['DEST','QTY'],axis = 1)

amphur_miss_ph = amphur_miss_ph.join([one_hot_d])

amphur_miss_ph
```

```python
# for traning

amphur_nonmiss_ph = phrae.loc[phrae['re_AMPHUR'].notnull()]

one_hot_d = pd.get_dummies(amphur_nonmiss_ph['DEST'], prefix='', prefix_sep='',dtype=int)

amphur_nonmiss_ph = amphur_nonmiss_ph.drop(['DEST','QTY'],axis = 1)

amphur_nonmiss_ph = amphur_nonmiss_ph.join([one_hot_d])

amphur_nonmiss_ph
```

```python
missing_columns__nonmiss = set(amphur_nonmiss_ph.columns) - set(amphur_miss_ph.columns)
print(missing_columns__nonmiss)
for j in missing_columns__nonmiss:
    amphur_miss_ph[j] = 0 
```

```python
missing_columns_miss = set(amphur_miss_ph.columns) - set(amphur_nonmiss_ph.columns)
print(missing_columns_miss)
for j in missing_columns_miss:
    amphur_nonmiss_ph[j] = 0 
```

```python
amphur_miss_ph = amphur_miss_ph.reindex(sorted(amphur_miss_ph.columns), axis=1)
amphur_nonmiss_ph = amphur_nonmiss_ph.reindex(sorted(amphur_nonmiss_ph.columns), axis=1)
amphur_miss_ph.columns.intersection(amphur_nonmiss_ph.columns) == amphur_nonmiss_ph.columns
```

```python
# Model

ph_x = amphur_nonmiss_ph.iloc[:,0:-2]
ph_y = amphur_nonmiss_ph.iloc[:,-2:-1]
ph_x_train, ph_x_test, ph_y_train, ph_y_test = train_test_split(ph_x, ph_y, test_size=0.75, stratify=ph_y)
model_ph = KNeighborsClassifier()
model_ph.fit(ph_x_train, ph_y_train)
ph_y_pred = model_ph.predict(ph_x_test) 
accuracy = accuracy_score(ph_y_test, ph_y_pred) 
print(accuracy)
```

```python
# oos_df[["Summer", "Winter", "Spring","Autumn"]].apply(lambda x: season_model(x),axis=1)
ph_x_miss = amphur_miss_ph.iloc[:,0:-2]
ph_y_pred_miss = model_ph.predict(ph_x_miss) 
ph_y_pred_miss
```

```python
amphur_miss_ph.re_AMPHUR = ph_y_pred_miss
amphur_miss_ph
```

```python
amphur_miss_ph.re_AMPHUR.isnull().sum()
```

```python
cleaned_ph = pd.concat([amphur_miss_ph,amphur_nonmiss_ph])
cleaned_ph.sort_index(ascending=True, inplace=True)
qty_ph = phrae.QTY.sort_index(ascending=True)
cleaned_ph = pd.merge(cleaned_ph, qty_ph,on='BILL_INDEX')
phrae = cleaned_ph.iloc[:,-3:]
phrae
```

```python
phrae_g = pd.DataFrame(phrae.groupby(by= [phrae.re_PROVINCE, phrae.re_AMPHUR])["QTY"].count())
phrae_g
```

```python
phrae.re_AMPHUR.replace("เมือง","เมืองแพร่", inplace=True)
phrae_g = pd.DataFrame(phrae.groupby(by= [phrae.re_PROVINCE, phrae.re_AMPHUR])["QTY"].count())
phrae_g
```
