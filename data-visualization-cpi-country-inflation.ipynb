```python
def country_plot(country,plotcolor="blue"):
    #Data Preprocessing
    df=pd.read_csv('/Users/phutharesuanwachirapisut/Programming/Dataset/global_inflation_data.csv')
    df=df.set_index("country_name")
    cntry=df.loc[[country]].iloc[:,1:]    
    
    #Data Visualization
    plt.style.use("dark_background")
    plt.figure(figsize=(16,6))
    plt.xticks(rotation=60) 
    plt.grid(False)
    plt.title(f"{country} CPI")
    plt.xlabel("Years")
    plt.ylabel("Inflation Rate")
    plt.plot(cntry.columns,cntry.loc[country].values,color=plotcolor)
    plt.show()
```

```python
country_plot('Australia',plotcolor="tomato")
```

```python
country_plot('Germany',plotcolor="lightblue")
```

```python
#Compare countries by CPI

def countries_plot(countries, plot_colors=None):
    # Data Preprocessing
    df = pd.read_csv('/Users/phutharesuanwachirapisut/Programming/Dataset/global_inflation_data.csv')
    df = df.set_index("country_name")
    
    # Data Visualization
    plt.style.use("dark_background")
    plt.figure(figsize=(16, 6))
    plt.xticks(rotation=60) 
    plt.grid(False)
    plt.title("CPI for Countries")
    plt.xlabel("Years")
    plt.ylabel("Inflation Rate")

    # If plot_colors is not provided, use default colors
    if plot_colors is None:
        plot_colors = ["blue", "orange", "green", "red", "purple", "cyan", "magenta", "yellow"]

    # Iterate over countries and plot their data
    for i, country in enumerate(countries):
        cntry = df.loc[[country]].iloc[:, 1:]
        color = plot_colors[i % len(plot_colors)]  # Cycle through colors if more countries than colors
        plt.plot(cntry.columns, cntry.loc[country].values, color=color, label=country)

    plt.legend()  # Show legend with country names
    plt.show()

```

```python
countries = ["United Kingdom", "Germany", "United States", "China, People's Republic of"]
colors = ["blue", "orange", "green", "red"]  # Provide custom colors if needed
countries_plot(countries, colors)
```
